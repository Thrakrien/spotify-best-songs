```python
import pandas as pd
```

```python
def search_music(music):
    musica1 = sp.search(q=str(music),type='track')
    batata1 = pd.DataFrame.from_dict(musica1,orient='index')
    batata1 = batata1.explode('items')
    batata1 = batata1['items'][0]
    
    idzinho = batata1['id']

    audiozinho = sp.audio_features(idzinho)[0]
    
    selected_data = {
    'danceability': audiozinho['danceability'],
    'energy': audiozinho['energy'],
    'speechiness': audiozinho['speechiness'],
    'acousticness': audiozinho['acousticness'],
    'liveness': audiozinho['liveness'],
    'valence':audiozinho['valence']
    }

    
    doidera = pd.DataFrame(selected_data,index=[0])
    return doidera
```

```python
search_music('mc poze tropa do sabio')
```

```python
search_music('toe goodbye')
```

```python
loaded_model.predict(search_music('toe goodbye'))
```
